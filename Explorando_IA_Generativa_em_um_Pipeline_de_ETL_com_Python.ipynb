```python
import pandas as pd

df = pd.read_csv('Clientes_Id.csv')
df.head()
```

```python
import os

from groq import Groq

client = Groq(api_key=os.environ.get("GROQ_API_KEY"))


def generate_ai_news(nome):
  completion = client
  chat_completion = client.chat.completions.create(
  messages=[
        {
            "role": "system",
            "content": ("""Você é um especialista em marketing bancário.
            não retorne chain-of-thought leak
            proibido retornar chain-of-thought leak

            não retorne chain-of-thought leak e suas marcações
            proibido retornar chain-of-thought leak e suas marcações
            proibido incluir blocos internos ou marcações.
            sem incluir blocos internos ou marcações.

          .""")
        },
        {
            "role": "user",
            "content": f"Crie uma mensagem para o {nome}, em portugues brasileiro, sobre a immportancia dos investimentos no maximo de 100 caracteres."
        }
  ],
  temperature=0.4,
  top_p=0.9,
  max_tokens=60,
  frequency_penalty=0.2,
  presence_penalty=0.1,
  model="meta-llama/llama-4-maverick-17b-128e-instruct"
  #model="qwen/qwen3-32b",
  )
  return chat_completion.choices[0].message.content

```

```python
df.columns = df.columns.str.strip()

user_ids = df['UserID'].tolist()
print(user_ids)

for uid in user_ids:

    if uid in df['UserID'].values:
        nome = df.loc[df['UserID'] == uid, 'nome'].values[0]
        nova_msg = generate_ai_news(nome)

        df.loc[df['UserID'] == uid, 'mensagem'] = nova_msg

        print(f"ID {uid} → mensagem gerada:", nova_msg)

    else:
        print(f"ID {uid} não encontrado.")
```

```python
df.to_csv("/content/usuarios_atualizado.csv", index=False)
print("CSV salvo em /content/usuarios_atualizado.csv")
```
